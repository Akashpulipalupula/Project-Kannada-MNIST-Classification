```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
```

```python
# Extracting the dataset
X_train_dataset = np.load('E:\Data Science Course - GUVI\Final Project\Project - 1\Kannada_MNIST_datataset_paper\Kannada_MNIST_npz\Kannada_MNIST\X_kannada_MNIST_train.npz')
X_train = X_train_dataset['arr_0']
y_train_dataset = np.load('E:\Data Science Course - GUVI\Final Project\Project - 1\Kannada_MNIST_datataset_paper\Kannada_MNIST_npz\Kannada_MNIST\y_kannada_MNIST_train.npz')
y_train = y_train_dataset['arr_0']
X_test_dataset = np.load('E:\Data Science Course - GUVI\Final Project\Project - 1\Kannada_MNIST_datataset_paper\Kannada_MNIST_npz\Kannada_MNIST\X_kannada_MNIST_test.npz')
X_test = X_test_dataset['arr_0']
y_test_dataset = np.load('E:\Data Science Course - GUVI\Final Project\Project - 1\Kannada_MNIST_datataset_paper\Kannada_MNIST_npz\Kannada_MNIST\y_kannada_MNIST_test.npz')
y_test = y_test_dataset['arr_0']
```

```python
# Component sizes
component_sizes = [10, 15, 20, 25, 30]

for n_components in component_sizes:
    print(f"Component Size: {n_components}")

    # PCA
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train.reshape(-1, 28*28))
    X_test_pca = pca.transform(X_test.reshape(-1, 28*28))

    # Models
    models = {
        'Decision Trees': DecisionTreeClassifier(),
        'Random Forest': RandomForestClassifier(),
        'Naive Bayes': GaussianNB(),
        'K-NN': KNeighborsClassifier()
    }

    for model_name, model in models.items():
        model.fit(X_train_pca, y_train)
        y_pred = model.predict(X_test_pca)

        # Metrics
        precision = precision_score(y_test, y_pred, average='macro')
        recall = recall_score(y_test, y_pred, average='macro')
        f1 = f1_score(y_test, y_pred, average='macro')
        cm = confusion_matrix(y_test, y_pred)
        y_prob = model.predict_proba(X_test_pca)
        roc_auc = roc_auc_score(y_test, y_prob, multi_class='ovr')

        print(f"Metrics for {model_name}:")
        print(f"Precision: {precision:.4f}")
        print(f"Recall: {recall:.4f}")
        print(f"F1-Score: {f1:.4f}")
        print("Confusion Matrix:")
        print(cm)
        print(f"ROC-AUC Score: {roc_auc:.4f}")
        print("\n")

        # Plotting ROC curve
        fpr = {}
        tpr = {}
        roc_auc = {}

        for class_label in range(10):
            fpr[class_label], tpr[class_label], _ = roc_curve(y_test == class_label, y_prob[:, class_label])
            roc_auc[class_label] = roc_auc_score(y_test == class_label, y_prob[:, class_label])

        plt.figure()
        for class_label in range(10):
            plt.plot(fpr[class_label], tpr[class_label], label=f'Class {class_label}, AUC = {roc_auc[class_label]:.4f}')
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve for {model_name}')
        plt.legend(loc='lower right')
        plt.show()
```

```python
# SVM
# Component sizes
component_sizes = [10,20,30]
```

```python
for n_components in component_sizes:
    print(f"Component Size: {n_components}")

    # PCA
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train.reshape(-1, 28*28))
    X_test_pca = pca.transform(X_test.reshape(-1, 28*28))

    # SVM
    svm = SVC(probability=True, decision_function_shape='ovr')
    svm.fit(X_train_pca, y_train)
    y_pred = svm.predict(X_test_pca)
    y_prob = svm.predict_proba(X_test_pca)

    # Metrics
    precision = precision_score(y_test, y_pred, average='macro')
    recall = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')
    cm = confusion_matrix(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_prob, multi_class='ovr')

    print("Metrics for SVM:")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print("Confusion Matrix:")
    print(cm)
    print(f"ROC-AUC Score: {roc_auc:.4f}")

    # Plotting ROC curve
    fpr = {}
    tpr = {}
    roc_auc = {}

    for class_label in range(10):
        fpr[class_label], tpr[class_label], _ = roc_curve(y_test == class_label, y_prob[:, class_label])
        roc_auc[class_label] = roc_auc_score(y_test == class_label, y_prob[:, class_label])

    plt.figure()
    for class_label in range(10):
        plt.plot(fpr[class_label], tpr[class_label], label=f'Class {class_label}, AUC = {roc_auc[class_label]:.4f}')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve for SVM')
    plt.legend(loc='lower right')
    plt.show()
```
